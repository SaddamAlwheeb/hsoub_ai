```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Bank_Marketing_Campaign.csv')
df
```

```python
df.isnull().sum()
```

```python
percentage_value_null = (df.isnull().sum()/df.shape[0])*100
percentage_value_null.sort_values(ascending=False)
```

```python
cat_cols = [df.columns[i] for i in range(1,df.shape[1]-1) if df.iloc[:,i].dtype=='O']
# print(df.iloc[:,2].dtype)
cat_cols
```

```python
fig, axes = plt.subplots(5, 2, figsize=(18, 30))
axes = [ax  for axes_rows in axes for ax in axes_rows]


for i,column in enumerate(df[cat_cols]):
    df[column].value_counts()[::-1].plot(kind='pie',ax=axes[i], title=column,autopct='%.0f%%', fontsize=12)


    
    
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 30))
axes = [ax  for axes_rows in axes for ax in axes_rows]


for i,column in enumerate(df[cat_cols]):
    df[column].value_counts()[::-1].plot(kind='barh',ax=axes[i], title=column)
```
