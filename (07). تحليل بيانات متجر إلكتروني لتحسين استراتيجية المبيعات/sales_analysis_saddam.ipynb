```python
df = frame[frame['quantity_ordered']!='Quantity Ordered']
df = df[df['price_each']!='Price Each']
```

```python
df = df.astype({"quantity_ordered": int, "price_each": float})
```

```python
df.info()
```

```python
df
```

```python
df['sales']=df['price_each']*df['quantity_ordered']
df
```

```python
product_sales= df.groupby('product')['sales'].agg(sales_sum='sum').sort_values(by='sales_sum',ascending=False).reset_index()
product_sales
```

```python
fix ,ax = plt.subplots(figsize=(16,4))
ax = sns.barplot(x= 'sales_sum',y='product',data= product_sales)

plt.title('product sales')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df
```

```python
def get_city(address):
    return address.split(",")[1].strip(' ')

def get_state(address):
    return address.split(",")[2].split(" ")[1]
    

df['city']=df['purchase_address'].apply(lambda x :  get_city(x) + " (" + get_state(x) + ")")
df
```

```python
city_sales = df.groupby('city')['sales'].agg(sum_sales='sum').sort_values(by = 'sum_sales',ascending=False).reset_index()
city_sales
```

```python
fig , ax = plt.subplots(figsize=(16,4))

ax = sns.barplot(x= 'city',y='sum_sales',data= city_sales)
plt.title('city plot')

ax.bar_label(ax.containers[0])
plt.show()
```

```python
time_sales = df
list_time = []
for i in time_sales['order_date']:
    list_time.append(i.split(" ")[1])
    
time_sales['time'] = list_time
time_sales.head()
```

```python
def get_time(date):
    return date.split(' ')[1]
df['time']=df['order_date'].apply(lambda x: get_time(x))
df.head()
```

```python
time_sales = df.groupby('time')['quantity_ordered'].agg(time_sales = 'sum').sort_values(by="time_sales", ascending=False).reset_index()
time_sales.head()
```

```python
fig , ax = plt.subplots(figsize=(16,4))

ax = sns.barplot(x= 'time_sales',y='time',data= df)
plt.title('time plot')

ax.bar_label(ax.containers[0])
plt.show()
```
